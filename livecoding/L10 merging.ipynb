```python
import pandas as pd
# this "firm" variable is really "last firm" but the database manager called it what they called it 
# and now IT is scared that fixing it will break all their systems, so it's "stuck as-is"
some_employees = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue',"Anderson"],
                    'firm': ['Citi','GS','Accenture','Wells Fargo',"Lehigh"],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR','Finance'],
                    'hire_date': [2004, 2008, 2012, 2014,2020]})

departments = pd.DataFrame({'group': ['Accounting', 'Engineering', 'HR',"Janitors"],
                    'employees': [100,200,1,2] })

current_employer = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'firm': ['Yahoo','Yahoo','Yahoo','Yahoo']})

some_employees_alt = pd.DataFrame({'employee': ['Jonathan', 'Mary', 'Lisa', 'Sue'],
                    'firm': ['?','Lehigh','Accenture','Wells Fargo'],
                    'group': ['Accounting', 'CEO', 'Engineering', 'HR'],
                    'hire_date': [2010, 2019, 2012, 2014]})
```

```python
### CONFLICTING NAMES? 2 OPTIONS

# specify suffix! x and y aren't informative
pd.merge(left=some_employees,right=current_employer,
        on='employee',how='left',suffixes=('_prior','_current'))

# OR

pd.merge(left=some_employees,right=current_employer,
        on='employee',how='left').rename(columns={'firm_x':'previous','firm_y':'now'})
```

## Missing values

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"A":[12, 4, 5, None, 1], 
                   "B":[None, 2, 54, 3, None], 
                   "C":[20, 16, None, 3, 8], 
                   "D":[14, 3, None, None, 6]}) 
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
df[['A','B']].dropna()
```

```python
# fill them
df.fillna(0)

df2 = df.copy()
df2['A'] = df2['A'].fillna(1)   # change just one variable
df2


df.fillna(df.mean())

df.fillna(method='bfill') # carries values backwards

df.interpolate()   # weird ... doesn't back fill but forward fills
```

## Those examples are dumb, no datasets are like that

```python
# create a firm level data set

df4 = df.copy()
df4['Firm']=1

df5 = df.copy()
df5['Firm']=2

df6=df4.append(df5)
df6
```

```python
# carry forward values

df6.groupby('Firm').fillna(method='ffill')
```
